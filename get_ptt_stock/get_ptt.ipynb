```python
from  bs4 import BeautifulSoup as BS
import requests 
response = requests.get('https://www.ptt.cc/bbs/Stock/index1.html')
soup = BS(response.text, 'html.parser') 
print(soup)
```

透過tag尋找

```python
print(soup.select('div.title a'))
```

```python
soup.select('div.topbar-container a')
```
